```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
```

```python
data={'name':['ahmad','ali','hind','amal','mansour','fasial','sara','ahmad','manar','fahad','nojoud'],
     'age':[32,40,10,34,55,40,20,15,55,15,18],
     'gender':['M','M','F','F','M','M','F','M','F','M','F'],
     'skill':['reading','neither','writing','writing','neither','writing','neither','writing','reading','reading',np.nan]}
df=pd.DataFrame(data=data)
```

```python
df
```

```python
encoder = LabelEncoder()
df['gender_encoded']=encoder.fit_transform(df['gender'])
```

```python
df
```

```python
x=df.drop(index=[10])
y=df.loc[10,:]
```

```python
x
```

```python
y
```

```python
def distance(age,gender):
    for i,row in x.iterrows():
        dis=np.sqrt(((age-row['age'])**2)+((gender-row['gender_encoded'])**2))
        x.loc[i,'distance']=dis
```

```python
distance(y['age'],y['gender_encoded'])
```

```python
x
```

```python
def predict(k,x):
    x = x.sort_values(by='distance').reset_index()
    skills = x.loc[0:k-1,'skill']
    skill=skills.value_counts().sort_values(ascending=False).index[0]
    return skill
```

```python
predict(1,x)
```
